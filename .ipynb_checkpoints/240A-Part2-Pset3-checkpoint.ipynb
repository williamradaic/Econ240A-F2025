# Problem 1 - Bayesian Bootstrap

```python
## (Based on section code) ##

## Setup and loading data


# Direct Python to plot all figures inline (i.e., not in a separate window)
%matplotlib inline

# Load libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load data
nlsy97 = pd.read_csv('data/nlsy97ss.csv')
# nlsy97['asvab'] = nlsy97['asvab'] / 10; # Normalize AFQT 
nlsy97.head()
```

```python
## 1.1 Create subsample 

# Subsample indicator
sample = (nlsy97['female']==0) & (nlsy97['black']==0) & (nlsy97['hispanic']==0) & (nlsy97['avg_earn_2014_to_2018']>0)
sample.head()

# Filtering dataset
nlsy97 = nlsy97[sample]

# Creating log(earnings)
nlsy97['LogEarn'] = np.log(nlsy97['avg_earn_2014_to_2018'])

# Creating interaction term (for item 6)
nlsy97['hgc_X_asvab'] = nlsy97['hgc_ever']*(nlsy97['asvab'] - 50)

# Summary statistics
nlsy97[["avg_earn_2014_to_2018","LogEarn","hgc_ever","asvab"]].describe()
```

```python
## 1.2. OLS fit - short regression: LogEarn ~ 1 + hgc_ever

short_reg=sm.OLS(nlsy97['LogEarn'],sm.add_constant(nlsy97['hgc_ever'])).fit(cov_type='HC0') # HC0 is heteroskedastic-robust s.e.
short_reg.summary()
```

```python
## 1.3. Long regression: LogEarn ~ 1 + hgc_ever + asvab

long_reg=sm.OLS(nlsy97['LogEarn'],sm.add_constant(nlsy97[['hgc_ever','asvab']])).fit(cov_type='HC0')
long_reg.summary()
```

Yes, the estimate on `hgc_ever` goes down from 0.0973 to 0.0732. This is because schooling and "ability" are correlated; in the short regression, the coefficient associated with schooling is picking up some of the effect of ability. This is usually called "Ommited Variable Bias" in the literature.

```python
## 1.4. Auxiliary regression: asvab ~ 1 + hgc_ever

auxi_reg=sm.OLS(nlsy97['asvab'],sm.add_constant(nlsy97['hgc_ever'])).fit(cov_type='HC0')
auxi_reg.summary()
```

**1.5.** We use the FWL theorem to reconstruct the long-regression coefficient:
$$ \delta = \beta + \Pi \gamma $$
where $\delta$ is the short-regression coefficient for `hgc_ever`; $\beta$ is the long-regression coefficient for `hgc_ever`; $\gamma$ is the long-regression coefficient on `asvab`; and $\Pi$ is the auxiliary-regression coefficient for `hgc_ever`.

```python
long_reg_FWL = long_reg.params['hgc_ever']+auxi_reg.params['hgc_ever']*long_reg.params['asvab']
print("delta =", long_reg_FWL)
```

By the FWL theorem, the coefficient should be equal (up to rounding/floating point errors). Checking this below:

```python
np.isclose(long_reg.params['hgc_ever']+auxi_reg.params['hgc_ever']*long_reg.params['asvab'],short_reg.params['hgc_ever'])
```

```python
## 1.6. Including interaction term in the long regression

inter_reg=sm.OLS(nlsy97['LogEarn'],sm.add_constant(nlsy97[['hgc_ever','hgc_X_asvab','asvab']])).fit(cov_type='HC0')
inter_reg.summary()
```

**(a)** The interpretation of $\beta_0$ is: for an individual with AFQT score of 50, all else constant, an increase of 1 year of schooling increases earnings by 7% on average.

**(b)** All else constant, an additional year of schooling leads to an increase of $\beta_0 + \gamma_0 (asvab - 50)\% = (7.13 + 0.03(asvab - 50))\%$ in earnings. This corresponds to the direct average effect of education on earnings combined with the interaction effect of education with ability.

```python
## 1.7. Plot

# get OLS estimate
beta0, gamma0 = inter_reg.params[['hgc_ever','hgc_X_asvab']]
x_grid = np.linspace(min(nlsy97["asvab"]), max(nlsy97["asvab"]), 100)
y = beta0 + gamma0 * (x_grid - 50)
# construct a plot
plt.figure(figsize=(8,6))
plt.plot(x_grid, y)
plt.xlabel("asvab")
plt.ylabel("$\\beta_0 + \\gamma_0$(asvab-50)")
plt.ylim([0.05, 0.1]);
```

```python
## 1.8. Bayes Bootstrap

y = nlsy97['LogEarn']
X = sm.add_constant(nlsy97[['hgc_ever','hgc_X_asvab','asvab']]) # "Permanently" add constant to design matrix
S = 1000 # Number of posterior draws
M = np.empty((S,101)) # Matrix with posterior draws of statistics of interest
N = len(y) # Number of observations in dataset

for s in range (0,S):
    W = np.random.gamma(1.,1.,N) # Random draws of Gamma(1,1) variables
    W = W/np.sum(W) # Converting draws to Dirichlet
    result = sm.WLS(y,X,weights=W).fit()
    M[s,:] = result.params['hgc_ever']+result.params['hgc_X_asvab']*(np.array(range(0,101)) - 50) # Linear regression with Dirichlet wgts

plq = np.quantile(M, 0.025, axis=0)
puq = np.quantile(M, 0.975, axis=0)
pm = np.mean(M, axis=0)

## Plot
fig = plt.figure(figsize=(8, 6))
ax = plt.axes()

ax.fill_between(x=np.array(range(0,101)), y1=plq, y2=puq, color='#FDB515',alpha=0.5)
ax.plot(np.array(range(0,101)), pm, ',-', color="#003262")

ax.set_title('Posterior distribution of schooling returns')
ax.set_xlabel('asvab percentile')
ax.set_ylabel('return-to-schooling')

classes = ['0.025 to 0.975 inter-quantile range','posterior mean']
plt.legend(labels=classes)

plt.ylim([0.01,0.15])
plt.xlim([0,100])

# Clean up the plot, add frames etc.
ax.patch.set_facecolor('gray') # Color of background
ax.patch.set_alpha(0.15) # Translucency of background
ax.grid(False)

# Add frame around plot
for spine in ['left','right','top','bottom']:
    ax.spines[spine].set_visible(True)
    ax.spines[spine].set_color('k')
    ax.spines[spine].set_linewidth(2)

# Save plot
plt.tight_layout()
plt.show()
```

**1.9.** In the subsample at hand, we see that: (i) earnings is positively correlated with education; (ii) AFQT, a proxy for "ability", is correlated with both earnings and schooling; (iii) as the plot above shows, people with higher AFQT scores have higher returns to schooling; (iv) the estimates are more uncertain at the tails of the distribution of AFQT.

# Problem 2 - Metropolis-Hastings

**2.1.** The Metropolis-Hastings (MH) algorithm is a computationally efficient procedure to approximate posterior distributions when exact computations are intractable. 
In the problem at hand, we have specified prior subjective uncertainty over the parameters $\phi$ and $\sigma^2$, and want to use the data to compute the posterior distribution of these parameters.
In order to construct this posterior distribution, we need to approximate more draws from this posterior, since the observed data corresponds to just a single draw.
The key reason why the procedure works is that the Markov Chain induced by the algorithm converges to a stationary distribution as $s$ grows---i.e., converges to the posterior distribution. Thus, we can approximate draws from the posterior and construct its distribution.
$\theta'$ is always accepted whenever $\alpha > 1$---i.e., whenever the likelihood ratio for the new parameter is larger than 1. This means that the new parameter $\theta'$ is an improvement (in terms of likelihood) relative to $\theta$. If $\alpha < 1$, the new parameter will be adopted with positive probability given by the CDF of a Uniform(0,1) evaluated at $\alpha$; that is, whenever $\theta'$ leads to a reduction in likelihood, it can still be adopted with some positive probability. The reasoning behind this decision rule is that it allows for the algorithm to visit all the parameter space and not get stuck in local minima.

**2.2.**

**(a)** 
$\kappa_\phi$ scales the variance of the distribution $\phi'\mid \theta$, from which new values of $\phi$ will be drawn. A higher value of $\kappa_\phi$ corresponds to a more dispersed distribution, which means draws $\phi'$ are less likely to be concentrated near the old parameter $\phi$.
$\kappa_\sigma$ scales the shape and the scale of the distribution of new values of $\sigma^{2\prime} \mid \theta$. Again, a higher value of $\kappa_\phi$ corresponds to a more dispersed distribution, which means draws $\sigma^{2\prime}$ are less likely to be concentrated near the old parameter $\sigma^2$. In other words, increasing $\kappa$ in the implementation of the algorithm leads to a more dispersed search process, visiting more regions of the parameter space with higher frequency. In terms of the Markov Chain, it means the transition matrix will be less sparse, in the sense of having more equally distributed values across the entire matrix.

**(b)** 
Since $\phi'$ and $\sigma^{2\prime}$ are drawn indepdendently, the density $g(\theta' \mid \theta)$ is equal to 
$$ g(\theta' \mid \theta) = g(\phi' \mid \theta) g(\sigma^{2\prime} \mid \theta) $$
where $g(\phi' \mid \theta)$ and $g(\sigma^{2\prime} \mid \theta)$ are the pdfs associated with $\phi' \mid \theta$ and $\sigma^{2\prime} \mid \theta$ respectively.

Since $\phi'\mid \theta \sim \mathcal{N}(\phi, \kappa_\phi I_k)$, we can write for the case of the 4th and 5th draws:
$$g(\phi^{(5)} \mid \theta^{(4)}) = (2 \pi)^{-k / 2} \kappa_\phi^{-k / 2} \exp \left(-\frac{1}{2}(\phi^{(5)}-\phi^{(4)})^{\mathrm{T}} \kappa_\phi^{-1} (\phi^{(5)}-\phi^{(4)})\right).$$

Similarly, for $\sigma^{2\prime} \mid \theta$ in the case at hand we have:
$$
g(\sigma^{2(5)} \mid \theta^{(4)}) 
= \frac{
\kappa_\sigma^{\,\kappa_\sigma \sigma^{2(4)}}
}{
\Gamma(\kappa_\sigma \sigma^{2(4)})
}
\,
(\sigma^{2(5)})^{\kappa_\sigma \sigma^{2(4)} - 1}
\exp\!\left( -\kappa_\sigma \sigma^{2(5)} \right).$$




Multiplying both expressions above gives us the final expression for $g(\theta' \mid \theta)$:
$$g(\theta^{(5)} \mid \theta^{(4)}) = (2 \pi)^{-k / 2} \kappa_\phi^{-k / 2} \exp \left(-\frac{1}{2}(\phi^{(5)}-\phi^{(4)})^{\mathrm{T}} \kappa_\phi^{-1} (\phi^{(5)}-\phi^{(4)})\right) \frac{
\kappa_\sigma^{\,\kappa_\sigma \sigma^{2(4)}}
}{
\Gamma(\kappa_\sigma \sigma^{2(4)})
}
\,
(\sigma^{2(5)})^{\kappa_\sigma \sigma^{2(4)} - 1}
\exp\!\left( -\kappa_\sigma \sigma^{2(5)} \right)$$


**(c)** 
Since the normal pdf is symmetric, the terms related to $\phi \mid \theta$ cancel out. Thus, this correction term is just a ratio of the Gamma distributions. After algebra, we get:

$$\frac{g(\theta \mid \theta')}{g(\theta' \mid \theta)}
=
\frac{\Gamma\!\big(\kappa_\sigma \sigma^{2(4)}\big)}
     {\Gamma\!\big(\kappa_\sigma \sigma^{2(5)}\big)}
\,
\kappa_\sigma^{\,\kappa_\sigma \big(\sigma^{2(5)} - \sigma^{2(4)}\big)}
\,
\frac{
\big(\sigma^{2(4)}\big)^{\kappa_\sigma \sigma^{2(5)} - 1}
}{
\big(\sigma^{2(5)}\big)^{\kappa_\sigma \sigma^{2(4)} - 1}
}
\,
\exp\!\left[-\kappa_\sigma\big(\sigma^{2(4)} - \sigma^{2(5)}\big)\right].$$

**2.3.**
We start by factoring the posterior ratio as 
$$
\frac{\pi(\theta' \mid Y)}{\pi(\theta \mid Y)}
=
\frac{
p(Y \mid \phi',\sigma'^2)\,p(\phi' \mid \sigma'^2)\,p(\sigma'^2)
}{
p(Y \mid \phi,\sigma^2)\,p(\phi \mid \sigma^2)\,p(\sigma^2)
},$$
where the first term is the likelihood ratio; the second term is the (Normal) prior on $\phi$, and the third term is the Inverse-Gamma prior on $\sigma^2$.

Given that these are ratios, the constant terms in each of the pdfs cancel out. Thus, after algebra, we have:
$$
\frac{\pi(\theta' \mid Y)}{\pi(\theta \mid Y)}=
\left(\frac{\sigma'^2}{\sigma^2}\right)^{-\frac{n+K}{2}}
\left(\frac{\sigma^2}{\sigma'^2}\right)^{\alpha+1}
\exp\!\left[
  -\frac{1}{2\sigma'^2}\big(\|Y - X\phi'\|^2 + (\phi' - \bar\phi)'\Lambda(\phi' - \bar\phi)\big)
  +\frac{1}{2\sigma^2}\big(\|Y - X\phi\|^2 + (\phi - \bar\phi)'\Lambda(\phi - \bar\phi)\big)
  -\frac{\beta}{\sigma'^2} + \frac{\beta}{\sigma^2}
\right]
$$
where $(\alpha = \nu_0/2,\ \beta = \nu_0 s_0^2/2)$.
Interpreting the expression above: $\|Y - X\phi'\|^2$ and $\|Y - X\phi\|^2$ come from the LR (with no Var-Cov matrix since we assume homoskedasticity); $(\phi' - \bar\phi)'\Lambda(\phi' - \bar\phi)\big)$ come from the prior on $\phi$, with associated Var-Cov matrix $\Lambda$; and the $\beta$ fractions come from the Inverse-Gamma prior.

**2.4.**
**(a)** The hyperparameters determine our prior subjective uncertainty regarding the parameters of interest $\phi$. $\bar{\phi}$ is the mean of the prior (Normal) distribution: it represents the average value of $\phi$ we have in our heads before conducting the analysis. $\Lambda^{-1}$ represents the covariances between different dimensions of the parameter $\phi$. For example, we may expect the effects of schooling and AFQT on earnings to be correlated, such that the prior uncertainty for $\phi$ carries this correlation structure. $v_0$ and $s_0^2$ determine the shape and scale of the Inverse-Gamma distribution. Larger values of $s_0$ correspond to a more concentrated prior on variance $\sigma^2$; larger values of $v_0$ (focusing on its effect on shape) shift more density to lower values of $\sigma^2$. Thus: large $v_0$ means that our prior is that $U$ has low variance; high $s_0$ means our prior itself is more concentrated.

**(b)** When $S$ is large, we approach the stationary distribution of the Markov Chain --- which is unique due to ergodicity. Thus for large $S$, as $s$ increases, the prior will play a negligible role in the evolution of the Markov Chain.

```python
## 1.5. 

# Load data
nlsy97 = pd.read_csv('data/nlsy97ss.csv')
# nlsy97['asvab'] = nlsy97['asvab'] / 10; # Normalize AFQT 
nlsy97.head()

## 1.1 Create subsample 

# Subsample indicator
sample = (nlsy97['female']==0) & (nlsy97['black']==0) & (nlsy97['hispanic']==0) & (nlsy97['avg_earn_2014_to_2018']>0)
sample.head()

# Filtering dataset
nlsy97 = nlsy97[sample]

# Creating log(earnings)
nlsy97['LogEarn'] = np.log(nlsy97['avg_earn_2014_to_2018'])

# Creating interaction term (for item 6)
nlsy97['hgc_X_asvab'] = nlsy97['hgc_ever']*(nlsy97['asvab'] - 50)

# Setup matrices and indices
Y = nlsy97["LogEarn"]
X = nlsy97[["hgc_ever", "hgc_X_asvab", "asvab"]]
X.insert(0, 'constant', 1)
n, K = X.shape

X.head()
```

```python
##### Defining functions:

# ---------------------------------
# 1. Log-likelihood  p(Y | , )
# ---------------------------------

def log_likelihood(phi, sig2):
    """
    Y | ,  ~ N(X,  I_N)
    """
    if sig2 <= 0:
        return -np.inf
    resid = Y - X @ phi
    rss = resid @ resid          # ||Y - X||
    return -0.5 * N * np.log(2 * np.pi * sig2) - 0.5 * rss / sig2


# ------------------------------------------------
# 2. Log-prior for  |   ~ N(,  ^{-1})
# ------------------------------------------------

def log_prior_phi(phi, sig2):
    """
     |  ~ N(,  ^{-1})
    Using  as a precision matrix.
    """
    if sig2 <= 0:
        return -np.inf
    diff = phi - phi_bar
    quad = diff @ (Lambda @ diff)      # ( - )'  ( - )
    # pdf of N_K(,  ^{-1})
    return (
        -0.5 * K * np.log(2 * np.pi * sig2)
        + 0.5 * logdet_Lambda
        - 0.5 * quad / sig2
    )


# ---------------------------------------------------
# 3. Log-prior for   ~ Inverse-Gamma(v0/2, v0 s0/2)
# ---------------------------------------------------

def log_prior_sig2(sig2):
    """
     ~ IG(, ) with density
    p()  ^ / () * ()^-(+1) * exp(- / )
    """
    if sig2 <= 0:
        return -np.inf

    alpha = v0 / 2.0
    beta  = v0 * s0_sq / 2.0

    return (
        alpha * np.log(beta)
        - lgamma(alpha)
        - (alpha + 1.0) * np.log(sig2)
        - beta / sig2
    )


# ------------------------------------------
# 4. Full log-posterior  ( | Y) up to const
#     = (, )
# ------------------------------------------

def log_posterior(phi, sig2):
    """
    (,  | Y)  p(Y | , ) p( | ) p()
    """
    return (
        log_likelihood(phi, sig2)
        + log_prior_phi(phi, sig2)
        + log_prior_sig2(sig2)
    )


# -----------------------------------------------------
# 5. Proposal kernel g(' | ) and Hastings correction
# -----------------------------------------------------

rng = np.random.default_rng(123)   # RNG for reproducibility


def propose(phi_curr, sig2_curr):
    """
    Draw ' = (', ') from the proposal:
      ' |  ~ N(, _ I_K)
      ' |  ~ Gamma(_ , _)  (shaperate parameterization)
    """
    # ' proposal (symmetric)
    phi_prop = phi_curr + np.sqrt(kappa_phi) * rng.normal(size=K)

    # ' proposal: Gamma(shape = _ , rate = _)
    shape = kappa_sigma * sig2_curr
    rate  = kappa_sigma
    # numpy.gamma uses shape and *scale*; scale = 1/rate
    sig2_prop = rng.gamma(shape=shape, scale=1.0 / rate)

    return phi_prop, sig2_prop


def _log_gamma_density(x, shape, rate):
    """
    Gamma(shape, rate) density:
    f(x) = rate^shape / (shape) * x^(shape-1) * exp(-rate x)
    """
    if x <= 0:
        return -np.inf
    return (
        shape * np.log(rate)
        - lgamma(shape)
        + (shape - 1.0) * np.log(x)
        - rate * x
    )


def log_proposal_ratio(phi_curr, sig2_curr, phi_prop, sig2_prop):
    """
    log g( | ') - log g(' | )

    Normal part for  is symmetric, so it cancels.
    Only the Gamma part in  contributes.
    """
    rate = kappa_sigma
    # forward: ' | _curr
    shape_fwd = kappa_sigma * sig2_curr
    # backward: _curr | '
    shape_bwd = kappa_sigma * sig2_prop

    return (
        _log_gamma_density(sig2_curr, shape_bwd, rate)
        - _log_gamma_density(sig2_prop, shape_fwd, rate)
    )
```

```python
def _log_gamma_density(x, shape, rate):
    """Gamma(shape, rate) density in log-space."""
    if x <= 0:
        return -np.inf
    return (
        shape * np.log(rate)
        - lgamma(shape)
        + (shape - 1.0) * np.log(x)
        - rate * x
    )


def propose_theta(phi_curr, sig2_curr, rng):
    """
    Draw ' = (', ') from the proposal:
        ' |   ~ N(, _ I_K)
        ' |  ~ Gamma(_ , _)
    """
    # ' proposal (symmetric)
    phi_prop = phi_curr + np.sqrt(kappa_phi) * rng.normal(size=phi_curr.shape)

    # ' proposal: Gamma(shape = _ , rate = _)
    shape = kappa_sigma * sig2_curr
    rate  = kappa_sigma
    sig2_prop = rng.gamma(shape=shape, scale=1.0 / rate)

    return phi_prop, sig2_prop


def log_proposal_ratio(phi_curr, sig2_curr, phi_prop, sig2_prop):
    """
    log g( | ') - log g(' | )

    Normal proposal for  is symmetric, so it cancels.
    Only the Gamma part in  contributes.
    """
    rate = kappa_sigma
    # forward: ' | _curr
    shape_fwd = kappa_sigma * sig2_curr
    # backward: _curr | '
    shape_bwd = kappa_sigma * sig2_prop

    return (
        _log_gamma_density(sig2_curr, shape_bwd, rate)
        - _log_gamma_density(sig2_prop, shape_fwd, rate)
    )


# ---------------------------------------------------------
# Main MetropolisHastings sampler
# ---------------------------------------------------------

def run_mh(S=20000, burn_in=5000, phi_init=None, sig2_init=None, seed=0):
    """
    Run MH for  = (, ).

    Returns a dict with:
      - 'phi': draws for  after burn-in (array of shape (S - burn_in, K))
      - 'sig2': draws for  after burn-in (array of shape (S - burn_in,))
      - 'phi_all', 'sig2_all': full chains including burn-in
      - 'accept_rate': overall acceptance rate
    """
    rng = np.random.default_rng(seed)

    N, K = X.shape

    # --- starting values: OLS + residual variance if not supplied ---
    if phi_init is None or sig2_init is None:
        beta_ols, *_ = np.linalg.lstsq(X, Y, rcond=None)
        resid = Y - X @ beta_ols
        sig2_ols = np.mean(resid ** 2)

        if phi_init is None:
            phi_init = beta_ols
        if sig2_init is None:
            sig2_init = sig2_ols

    phi_curr = np.asarray(phi_init, float)
    sig2_curr = float(sig2_init)

    # --- storage ---
    phi_chain = np.zeros((S + 1, K))
    sig2_chain = np.zeros(S + 1)

    phi_chain[0] = phi_curr
    sig2_chain[0] = sig2_curr

    accept = 0

    # --- MH iterations ---
    for s in range(S):
        phi_curr = phi_chain[s]
        sig2_curr = sig2_chain[s]

        # proposal
        phi_prop, sig2_prop = propose_theta(phi_curr, sig2_curr, rng)

        # log posterior at current and proposed
        log_post_curr = log_posterior(phi_curr, sig2_curr)
        log_post_prop = log_posterior(phi_prop, sig2_prop)

        # Hastings correction
        log_corr = log_proposal_ratio(phi_curr, sig2_curr, phi_prop, sig2_prop)

        # log acceptance probability: log  = min(0, log (') + log g(|') - log () - log g('|))
        log_alpha = (log_post_prop + log_corr) - log_post_curr
        if log_alpha > 0:
            log_alpha = 0.0  # min(0, ...) in log space

        # accept / reject
        if np.log(rng.uniform()) <= log_alpha:
            phi_chain[s + 1] = phi_prop
            sig2_chain[s + 1] = sig2_prop
            accept += 1
        else:
            phi_chain[s + 1] = phi_curr
            sig2_chain[s + 1] = sig2_curr

    accept_rate = accept / S

    # drop burn-in
    phi_post = phi_chain[burn_in + 1 :]
    sig2_post = sig2_chain[burn_in + 1 :]

    return {
        "phi": phi_post,
        "sig2": sig2_post,
        "phi_all": phi_chain,
        "sig2_all": sig2_chain,
        "accept_rate": accept_rate,
    }
```

```python
### Hyperparameters

import numpy as np
from math import lgamma

# ----------------------------
# Basic dimensions & priors
# ----------------------------

# X: (N, K), Y: (N,)
X = np.asarray(X, float)
Y = np.asarray(Y, float)
N, K = X.shape

# OLS estimate
beta_ols, *_ = np.linalg.lstsq(X, Y, rcond=None)
resid = Y - X @ beta_ols

# ---- Hyperparameters ----
# Prior:   |  ~ N(_bar,  ^{-1})
#          ~ IG( = v0/2,  = v0 * s0 / 2)

# Priors:
phi_bar = beta_ols                     # prior mean 
Lambda  = np.eye(K)                    # prior precision                 
v0      = 5.0                          # degrees-of-freedomtype hyperparameter
s0_sq = np.mean(resid**2)              # this is the prior scale for ^2

# Precompute stuff used many times
sign_L, logdet_Lambda = np.linalg.slogdet(Lambda)
assert sign_L > 0, " must be positive definite"
```

```python
### Run MH

j = 1            # coefficient on hgc_ever
S = 200000
burn_in = 0

# ---- Low variance chain ----
kappa_phi   = 1e-10      # small step size
kappa_sigma = 200       # relatively tight around current 

res_low = run_mh(S=S, burn_in=burn_in, seed=123)
beta_low = res_low["phi_all"][:, j]
print("Acceptance rate, low variance:", res_low["accept_rate"])
# ---- High variance chain ----
kappa_phi   = 5e-9      # bigger step
kappa_sigma = 100       # more dispersed  proposal

res_high = run_mh(S=S, burn_in=burn_in, seed=456)
beta_high = res_high["phi_all"][:, j]
print("Acceptance rate, high variance:", res_high["accept_rate"])
```

```python
## Plot figure of beta convergence

import matplotlib.pyplot as plt
import numpy as np

s_grid = np.arange(beta_low.shape[0])  # 0..S

fig, ax = plt.subplots(figsize=(8, 6))

burn_in_low = 125000
burn_in_high = 50000

ax.plot(s_grid, beta_low,  color="C0",  lw=0.7, label="Low")
ax.plot(s_grid, beta_high, color="red",   lw=0.7, label="High")

# # vertical dashed lines (here at burn-in, you can move them if you want)
# ax.axvline(burn_in_low, color="C0",  ls="--", alpha=0.8)
# ax.axvline(burn_in_high, color="red",   ls="--", alpha=0.8)

ax.set_title(r"Metropolis-Hastings Markov Chains for $\beta^{(s)}$")
ax.set_xlabel(r"Sample $s$")
ax.set_ylabel(r"$\beta^{(s)}$")
# plt.xlim(0, )
plt.ylim(0.05, 0.10)

leg = ax.legend(title="Variance", loc="lower right")
leg._legend_box.align = "left"   # pretty legend alignment

plt.tight_layout()
plt.show()
```

My prior is informed by the previous OLS analysis. Thus, I center the distributions at the OLS point estimates. Given the small number of observations in that analysis and the limited conclusions one may draw from that, though, I choose a dispersed distribution for $\phi$, with $\Lambda = I_K$.

As for both implementations of the MH algorithm: the Low variance (blue above) uses $\kappa_\phi =  10^{-10}$ and $\kappa_\sigma = 200$. This means that steps are really small and the variance distribution is pretty concentrated.
The High variance (red) used $\kappa_\phi = 5 \cdot 10^{-9}$ and $\kappa_\sigma = 100$. The higher $\kappa_\phi$ means that steps are bigger across iterations; and a lower $\kappa_\sigma$ makes the variance less concentrated. Finally, I use $S = 200,000$ iterations.

In the plot we see that the prior is influencing the estimates quite a bit. The low-variance algorithm has remarkably low movement around the OLS point estimate. The high-variance condition creates more movement, but still very low in magnitude --- variation is in the range of around 0.01. 
This highlights the importance of having some variance for the MH algorithm to work well. The procedure has to visit many regions of the parameter space, and having a prior that is too concentrated inhibits that from happening.

```python
## 1.6.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from math import lgamma, pi



# --------------------------------------------------
# 1. Choose which coefficient _j to plot
# --------------------------------------------------
j = 1  # for example, second coefficient

# posterior draws for _j from HIGH-variance chain (after burn-in)
beta_post = res_high["phi"][10000:, j] # 10000 burn in

# --------------------------------------------------
# 2. Posterior density via KDE
# --------------------------------------------------
kde = gaussian_kde(beta_post)

x_min, x_max = np.quantile(beta_post, [0.001, 0.999])
pad = 0.05 * (x_max - x_min)
x_grid = np.linspace(x_min - pad, x_max + pad, 400)

post_pdf = kde(x_grid)

# --------------------------------------------------
# 3. Prior density for _j
#     |  ~ N(,  ^{-1}),   ~ IG(v0/2, v0 s0 / 2)
# => marginal _j ~ Student-t_v0( mean = _j,
#                                scale = s0 * (^{-1})_jj )
# --------------------------------------------------
Lambda_inv = np.linalg.inv(Lambda)
df   = v0
mu   = phi_bar[j]
scale = np.sqrt(s0_sq * Lambda_inv[j, j])

def student_t_pdf(x, df, loc, scale):
    x = np.asarray(x)
    z = (x - loc) / scale
    log_c = (
        lgamma((df + 1) / 2)
        - lgamma(df / 2)
        - 0.5 * np.log(df * pi)
        - np.log(scale)
    )
    return np.exp(log_c) * (1 + z**2 / df) ** (-(df + 1) / 2)

prior_pdf = student_t_pdf(x_grid, df, mu, scale)

# --------------------------------------------------
# 4. Plot: histogram + posterior KDE + prior
# --------------------------------------------------
fig, ax = plt.subplots(figsize=(8, 6))

# posterior histogram
ax.hist(
    beta_post,
    bins=30,
    density=True,
    color="red",
    alpha=0.25,
    edgecolor="black"
)

# posterior KDE
ax.plot(x_grid, post_pdf, color="red", lw=2, label="Posterior density")

# prior density (often almost flat over this narrow x-range)
ax.plot(x_grid, prior_pdf, color="C0", lw=2, label="Prior density")

ax.set_title("Prior vs. Posterior Densities (high-variance MH)")
ax.set_xlabel(r"$\beta^{(s)}$")
ax.set_ylabel("Density")
plt.xlim(0.06, 0.1)
# plt.ylim(0, 0.3)

ax.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

The prior is pretty diffuse, since it follows $N(0,1)$. In comparison, the posterior is much more concentrated around 0.075. We can see the prior in blue in the figure above, and it is basically a line at zero.

**2.8**
In Problem 1, the prior distributions we use / assume only affect the results through varying the weights we give to observations in the bootstrap procedure. Thus, the bootstrap procedure itself in problem 1 is non-parametric. In Problem 2, on the other hand, we are making explicit assumptions regarding the distributions of the economic model at hand with normal errors. More importantly, it posits a specific structure to the *parameter uncertainty*, parametrizing it in terms of distributions (Normal for $\phi$, Inv-Gamma for $\sigma^2$). This parametrization guides the results, which are expressed in terms of posteriors for the parameters. In this setting, then, the uncertainty we capture is related to the parameters, not the empirical distribution.

In a broader sense (and depending on the interpretation you give it), Problem 1 can still be thought of as parametric, since it is imposing a linear structure to the variables at hand. If you treat the estimates as causal (say), then this also brings it closer to a parametric procedure.
